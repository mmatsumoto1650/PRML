```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,1,100)
x
```

```python
def model(x):
    return np.sin(2*np.pi*x)

t = model(x)
t
```

```python
plt.plot(x,t,color= 'green')
```

```python
x_sample = np.linspace(0,1,10)
t_sample = model(x_sample)+np.random.randn(10)*0.5
t_sample
```

```python
plt.plot(x,t,color= 'green')
plt.scatter(x_sample,t_sample)
```

```python
M = 10
def phi(x):
    return np.array([x**i for i in range(M+1)])

def poly(x,W):
    return phi(x).T @ W
```

```python
w_ = t_sample @ np.linalg.pinv(phi(x_sample))
w_
```

```python
plt.scatter(x_sample,t_sample)
plt.plot(x,t,color= 'green')
plt.plot(x,poly(x,w_))
```

```python
L = 0.005
w_2 = np.linalg.pinv(phi(x_sample) @ phi(x_sample).T + L*np.eye(M+1)) @ phi(x_sample) @ t_sample
w_2
```

```python
plt.plot(x,t,color= 'green')
plt.scatter(x_sample,t_sample)
plt.plot(x,poly(x,w_))
plt.plot(x,poly(x,w_2),color = 'red')
```
